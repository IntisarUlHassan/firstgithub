# Body Mass Index

## Body Mass Index (BMI)
- input weight
- height
- calculate BMI
- print BMI (my name is Intisar and my BMI is ----)

```python
weight=input('What is your mass? ')
```

```python
weight=int(weight)
```

```python
height=input('What is your height? ')
```

```python
height=float(height)
```

```python
name=input('what is your name? ')
```

```python
BMI= weight / (height**2)
```

```python
print('My name is', name, 'and my BMI is:',BMI, "k.g./m2")
```

# Fundamentals of Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)

titanic = sns.load_dataset("titanic")
sns.catplot(x="sex", y="survived", hue="class", kind='bar', data=titanic)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)
titanic = sns.load_dataset("titanic")
p1=sns.countplot(x="sex", hue="class", data=titanic)
p1.set_title("plot for counting")
plt.show()
```

```python
#scatter plot
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)
titanic = sns.load_dataset("titanic")
g=sns.FacetGrid(titanic, row="sex", hue="alone")
g=(g.map(plt.scatter, "age", "fare").add_legend())
plt.show()
```
